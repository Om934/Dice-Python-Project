# Dice Game - Mini project

```python
import random                                               # importing random module
import keyboard
def dice_game():                                            # defining a function
    
    
    ''' 1. First seek the input as players are ready to play
        2. Then seek player name as input
        3. rolling dice by the player and count score of a particular player based on chances he got as per problem statement.
        4. storing the data in a dictionary key as player name and score as value
        5. sorting the dictionary in decreasing order based on values
        6 . pick an first key and the display as winner   '''
    
    
    # code starts here
    
    players = int(input("Enter the no of players are ready to play dice game : ")) # seek an input from user
    d = {}                                                  # creating dictionary to store player as key and score as value
    for i in range(1 , players + 1):                        # using for loop to repeat the block of code as per input
        player_name = input(f"Enter player - {i} name : ")  # seek user name as input
        n = 5                                               # intialization chances varible  - n
        sum = 0                                             # intilization to store the score value for first time score starts with zero
        l = []                                              # creating empty list to store two consecutive elements
        while(n > 0):                                       # we using while loop as we dont know how many times we need to repeat the code
            n = n - 1                                       # updation of varible
            input("Plz Press Enter Key On Keyboard To Roll The Dice ...")
            num = random.randint(1,6)                      # using randint function to pick random from 1 - 6
            l.insert(0 , num)                               # inserting element at idex zero
            if len(l) > 3:                                  # if condition results boolean value
                l.pop()                                     # applying pop method to l list
            sum = sum + num                                 # we are calculating cummulative results of dice
            print("*"*50)                                   # print line
            print()                                         # giving line space
            print(f"Dice output for current play : {num} , chances left = {n}")
            print()                                         # giving line space
            if num == 6:                                    # if the dice output is 6 then we need to provide one more chance
                n = n + 1                                   # updation of varible n
                print(f"one more chance included due to you got {num} , chance left = {n}")
                print()                                     # giving line space
            if len(l) >= 2:
                if l[1] == l[0]:                            # to check weather two consecutive dice outputs are same
                    n = n + 1                               # updation of varible n
                    print(f"Your two consecutive dice outputs are same so , your get one extra chance , chances left = {n}")
                    print()                                 # Giving line space
            if len(l) >= 3:
                if l[0] == l[1] and l[1] == l[2]:           # checking three consecutive dice outputs are same 
                    n = 0                                   # making chances to Null to terminate the Game for that player.
                    print(f"Your three consecutive dice outputs are same , you lost all chances , chances left = {n}")
                    print()                                 # Giving line space
                    
                
                   
            print("*"*50)
        else:
            print()
            print("{} Score : {}".format(player_name , sum))
            print()
            if i != players:
                print('Coninue game with another player plz....')
            print()
            print("*"*50)
            d[player_name] = sum                            # adding dict values player names as keys and score as values
    d = dict(sorted(d.items() , key = lambda item : item[1] , reverse = True))# sorting a dictionary based on values in decreasing order
    for i,j in d.items():                                   # using for loop to get key and value in dictionary         
        print(f"{i} won the game with score : {j}")         # First key and corresponding value
        break                                               # using break statement to terminate for loop
        
        
    
        
        
dice_game()
```
